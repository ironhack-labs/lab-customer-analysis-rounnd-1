```python
# rearrange data and concatenate the 3 files in one data Frame
# go in df1 and change the name of the column

df1 = cust_data1
df2 = cust_data2
df3 = cust_data3

RG_data = pd.concat([df1,df2,df3], axis = 0 , join='outer')
RG_data
```

```python
# check for numerical columns
# wich columns are numerical, income , monthly_premium_auto       
# Which columns are categorical?        customer, st gender,education,customer_lifetime_value ,number_of_open_complaints, policy_type, vehicle_class, state , dtype


RG_data.dtypes
```

```python
# Delete the column education and the number of open complaints from the dataframe.

RG_data = RG_data.drop(['education','number_of_open_complaints'], axis = 1)
RG_data
```

```python
RG_data
```

```python
# Correct Data in customer_lifetime_value columns

Values_lifetime = RG_data["customer_lifetime_value"] 

corrected_value = pd.to_numeric(Values_lifetime, errors='coerce', downcast='float') * 100


corrected_value
```

```python
# check for duplicate row in the DATA and show them from the data Frame



duplicate_rows = RG_data[RG_data.duplicated]

duplicate_rows
```

```python
# now the Data frame with the rows duplcaite cleaned


No_Duplicate_rows = RG_data.drop_duplicates(subset="customer", keep='first', inplace=False, ignore_index=False)

No_Duplicate_rows

```

```python
# filter with a condition the customer with and income = or lower than 0

condition = (No_Duplicate_rows['income'] <= 0 )

List_income = No_Duplicate_rows[condition] 

List_income

```
