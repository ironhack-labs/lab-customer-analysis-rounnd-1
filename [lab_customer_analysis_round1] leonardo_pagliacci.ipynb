```python
import pandas as pd
```

```python
# Read the three files into python as dataframes

cus_data1 = pd.read_csv("/Users/leozinho.air/Desktop/Ironhack_DA/class_03/Lab 1 - customer analysis/lab-customer-analysis-round-1/files_for_lab/csv_files/file1.csv")

cus_data2 = pd.read_csv(r"/Users/leozinho.air/Desktop/Ironhack_DA/class_03/Lab 1 - customer analysis/lab-customer-analysis-round-1/files_for_lab/csv_files/file2.csv")

cus_data3 = pd.read_csv(r"/Users/leozinho.air/Desktop/Ironhack_DA/class_03/Lab 1 - customer analysis/lab-customer-analysis-round-1/files_for_lab/csv_files/file3.csv")


```

```python
# Show the DataFrame's shape.

print(cus_data1.shape)

print(cus_data2.shape)

print(cus_data3.shape)
```

```python
# Standardize header names.

cols1 = []

cols2 = []

cols3 = []


for i in range(len(cus_data1.columns)):
    cols1.append(cus_data1.columns[i].lower().replace(' ', '_'))

cus_data1.columns = cols1

cus_data1

for i in range(len(cus_data2.columns)):
    cols2.append(cus_data2.columns[i].lower().replace(' ', '_'))

cus_data2.columns = cols2

cus_data2

for i in range(len(cus_data3.columns)):
    cols3.append(cus_data3.columns[i].lower().replace(' ', '_'))

cus_data3.columns = cols3

cus_data3

    
```

```python
# Rearrange the columns in the dataframe as needed

# We use the columns of cus_data1 as the default columns for ours dataframes

cus_data2 = cus_data2[['customer', 'st', 'gender', 'education', 'customer_lifetime_value',
       'income', 'monthly_premium_auto', 'number_of_open_complaints',
       'policy_type', 'vehicle_class', 'total_claim_amount']] # Changing the order of the columns

cus_data2

# Now i am going to do the same for cus_data3

cus_data3['st'] = cus_data3['state']
cus_data3 = cus_data3.drop(columns ='state') # renaming the column 'state'

cus_data3 # Thankfully renamed

# Now let's order cus_data3

desired_order = ['customer', 'st', 'gender', 'education', 'customer_lifetime_value',
       'income', 'monthly_premium_auto', 'number_of_open_complaints',
       'policy_type', 'vehicle_class', 'total_claim_amount']
cus_data3 = cus_data3[desired_order]

# Checking if everything is ok
print(cus_data3.columns)
print(cus_data2.columns)
print(cus_data1.columns)
```

```python
# Concatenate the three dataframes

cus_data = pd.concat([cus_data1, cus_data2, cus_data3], axis = 0) # axis = 0 is not mandatory but is better putting it on

cus_data.tail(100)
```

```python
# Which columns are numerical?
# Which columns are categorial?

cus_data.dtypes


'''

'income','monthly_premium_auto','total_claim_amount ' are numerical, the other ones are object (strings or whatsoever).

customer                      object = Id referring to the customer
st                            object = State in which the car was bought
gender                        object = gender of the buyer
education                     object = education of the buyer
customer_lifetime_value       object = a metric that represents the total net profit a company can expect to generate from a customer throughout their entire relationship
income                       float64 = annual income of the buyer
monthly_premium_auto         float64 = monthly premium that the buyer has to pay
number_of_open_complaints     object = ...
policy_type                   object = ...
vehicle_class                 object = ...
total_claim_amount           float64 = ...

'''
```

```python
# Data cleaning

# Delete the column education and the number of open complaints from the dataframe

cus_data = cus_data.drop(['education'], axis = 1)

cus_data = cus_data.drop(['number_of_open_complaints'], axis = 1)

cus_data.head()
```

```python
# Correct the values in the column customer lifetime value. 
# They are given as a percent, so multiply them by 100 and change `dtype` to `numerical` type

cus_data = cus_data.dropna() # removing null values

cus_data['customer_lifetime_value']  = cus_data['customer_lifetime_value'].astype(str) # On cus_data3 (third df) the dtype of the customer lifetime column was 'float', then before cutting out the '%' character i had to convert the column to 'str' dtype

cus_data['customer_lifetime_value'] = cus_data['customer_lifetime_value'].str.replace('%','') # Excluding the '%' character from the colum so that i can convert all the column to float.

cus_data['customer_lifetime_value'] = cus_data['customer_lifetime_value'].astype(float) # Converting the column to float dtype.

cus_data['customer_lifetime_value'] = cus_data['customer_lifetime_value'].mul(100) # Multiplying the values by 100, the numbers are very large. ex -> 6.979536e+07 = 69,795,360


cus_data
```

```python
# Check for duplicate rows in the data and remove if any.

cus_data.duplicated()

cus_data.drop_duplicates()

```

```python
# Filter out the data for customers who have an income of 0 or less.

cus_data = cus_data.loc[cus_data['income'] > 0]

cus_data
```
