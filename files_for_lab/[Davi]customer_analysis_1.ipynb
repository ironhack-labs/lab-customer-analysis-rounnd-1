```python
import pandas as pd
import numpy as np
```

```python
# Read the three files into python as dataframes
df1 = pd.read_csv(r'files_for_lab\csv_files\file1.csv'), 
df2 = pd.read_csv(r'files_for_lab\csv_files\file2.csv'),
df3 = pd.read_csv(r'files_for_lab\csv_files\file2.csv')
```

```python
# Show the DataFrames shape (df1)
df1 = pd.read_csv(r'files_for_lab\csv_files\file1.csv')
df1.shape
```

```python
# Show the DataFrames shape (df2)
df2 = pd.read_csv(r'files_for_lab\csv_files\file2.csv')
df2.shape
```

```python
# Show the DataFrames shape (df3)
df3 = pd.read_csv(r'files_for_lab\csv_files\file3.csv')
df3.shape
```

```python
# Standardize header names.
df1.columns = df1.columns.str.lower().str.replace(' ', '_')
df2.columns = df2.columns.str.lower().str.replace(' ', '_')
df3.columns = df2.columns.str.lower().str.replace(' ', '_')

df3 = df3.rename(columns = {'state' : 'st'})
```

```python
df1_new_order = ['customer', 'gender', 'education', 'vehicle_class', 'st', 'income', 'policy_type', 'number_of_open_complaints', 'customer_lifetime_value', 'monthly_premium_auto', 'monthly_premium_auto', 'total_claim_amount']
order_df1 = df1[df1_new_order]
order_df1

df2_new_order = ['customer', 'gender', 'education', 'vehicle_class', 'st', 'income', 'policy_type', 'number_of_open_complaints', 'customer_lifetime_value', 'monthly_premium_auto', 'monthly_premium_auto', 'total_claim_amount']
order_df2 = df2[df2_new_order]

df3_new_order = ['customer', 'gender', 'education', 'vehicle_class', 'st', 'income', 'policy_type', 'number_of_open_complaints', 'customer_lifetime_value', 'monthly_premium_auto', 'monthly_premium_auto', 'total_claim_amount']
order_df3 = df3[df3_new_order]
```

```python
# Concatenate the three dataframes
customer_data = pd.concat([order_df1, order_df2, order_df3]).reset_index(drop=True)
customer_data  
```
