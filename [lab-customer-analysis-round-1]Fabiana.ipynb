# Data Transformation

```python
# Delete the column education and the number of open complaints from the dataframe.
merged_dfs = merged_dfs.drop(['education'], axis = 1)
```

```python
merged_dfs = merged_dfs.drop(['number_of_open_complaints'], axis = 1)
merged_dfs
```

```python
# Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change `dtype` to `numerical` type.
df['customer_lifetime_value'] = pd.to_numeric(df['customer_lifetime_value'].str.rstrip('%')) * 100
df['customer_lifetime_value']
```

```python
# Check for duplicate rows in the data and remove if any.
merged_dfs.duplicated()
merged_dfs.drop_duplicates()
```

```python
# Filter out the data for customers who have an income of 0 or less.
df = merged_dfs

filtered_df = df[['customer', 'income']][df['income'] <= 0]
filtered_df 
```
